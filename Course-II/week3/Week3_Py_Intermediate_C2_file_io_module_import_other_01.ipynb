```python
#week3-invoice-app-1-1.py

"""
-range, slicers, indexer, list comprehension
-main startup program
-Error Handling
-paramater input on proram startup
-file i/o
module import - custom python modules
"""

import datetime
import sys #python system library

#input data - collection - array/list/tuple/set/map
inv_data_1 = [89.34,78.23,54.24,33.324,14,123.29,31,26,51,23.12,123]

#create randmon/sample data

def calc_total():
  _total = 0.0
  _total = sum([item for item in inv_data_1])
  return _total

def save_invoice_data(username, email, total, items_total):
  f = open("invoice-data-1.txt","w")
  f.write("# Inoice System 1.0.3\n")
  f.write("\n")
  f.write(f"date: \t {username} \t {email} \n")
  f.write("\n")
  f.write(f"total:  {total} \n")
  f.write(f"items total:  {items_total} \n")
  f.close()

def main():

  #try/except - prevent code exit or code termination - by retieveing program errors/exceptions

  try:

    #print("file_name: ",sys.argv[0])
    username = sys.argv[1]
    email = sys.argv[2]

    _items_total = 0.0

    #red data from file
    f = open("items1.txt")

    for item in f:
      _items_total += float(item.strip())

    f.close()

    _total = calc_total()

    #display inovice app data
    print("# Inoice System 1.0.3")
    print()
    print("date:"," ", username, " ", email)
    print()
    print("total: ", _total)
    print("items total: ", _items_total)

    #save invoice data
    save_invoice_data(username,email,_total,_items_total)

  #retieve any type of error - Exception most generic/general error
  except Exception as error:
    print("program error")
    print(error)

#program entry  point or startup program - main
#c,c#java,c++ other
#main(String[] args)

if __name__ == "__main__":
  main()

#python program-name.py
#__name__ = "__main__"

#import program-name.py
#__name__ = "__module__"


```
