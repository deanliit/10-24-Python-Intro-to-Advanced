```python
#inv-demo-app-1.py
"""
# 1. Write a program to compute invoice/inventory items.

example:
item1 5
item2 10.5

total 15.5

"""

#input
#variables to store invoice items
#variable to store total
num1 = 5 #integer
num2 = 10.5 #float

#calculations
total = num1 + num2

#ouput
#display total output

#casting = covert from one data type to another data type

#print("total " + total) #error - compbine number to string
print("total " + str(total)) #convert number to string
#print("total",total,sep=":")
#print("total {}".format(total))
#print(f"total {total}")

```

```python
#inv-demo-app-2.py

#create variables

num1 = "5" #string
num2 = 10.5 #float - decimal places

#calculations
#total = num1 + num2 #error - convert num1 to integer
total = int(num1) + num2 #convert num1 to integer

#display output
print("total: " + str(total)) #casting/covert integer to string
#print("total",total,sep=":")
#print("total {}".format(total))
#print(f"total {total}")
```

```python
#inv-demo-app-3.py

num1 = "5" #string
num2 = "10.5" #string

#total = num1 + num2 #error - convert num1 to integer, num2 to float
#total = int(num1) + float(num2) #convert num1 to integer , num2 to float
total = float(num1) + float(num2) #convert num1 to integer , num2 to float

print("total: " + str(total)) #casting/covert integer to string
#print("total",total,sep=":")
#print("total {}".format(total))
#print(f"total {total}")
```
